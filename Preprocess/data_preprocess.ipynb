```python
import pandas as pd
import numpy as np
import datetime as dt 
import re
```

```python
twitter = pd.read_csv('../Data/data_elonmusk.csv', encoding = 'unicode_escape')
tsla = pd.read_csv('../Data/tsla_stock_price.csv')

```

```python
twitter.head()
```

```python
tsla.head()
```

```python
twitter.info()
tsla.info()
```

```python
twitter['Time'] = pd.to_datetime(twitter['Time'])
tsla['Date'] = pd.to_datetime(tsla['Date'])
twitter.info()
tsla.info()
```

```python
twitter['Date'] = twitter['Time'].dt.date
twitter['Date'] = pd.to_datetime(twitter['Date'])
print(twitter.head())
```

```python
twitter.sort_values(by=['Date'], inplace=True, ascending=True)
twitter.reset_index(drop=True, inplace=True)
twitter.head()
```

```python
# drop unwanted columns
twitter = twitter[['Tweet', 'Date']]
twitter.head()
```

```python
# merge tweets based on same date
i=0

for i, row in twitter.iterrows():
    stre = row['Tweet']
    
    # removing special characters
    new_string = re.sub('[^ a-zA-Z0-9]', '', stre)
    
    twitter.at[i, 'Date'] = row['Date']
    twitter.at[i, 'Tweet'] = new_string
    i += 1

twitter = twitter.groupby(['Date'], as_index=False).agg({'Tweet': ''.join}, inplace=True)
twitter.head()
```

```python
twitter.to_csv('twitter_cleaned.csv')
```

```python
merged = twitter.merge(tsla, left_on='Date', right_on='Date')
merged.head()
```

```python
# merged to csv
merged.to_csv('merged.csv')
merged['price_diff'] = merged['Close'] - merged['Open']
merged.head()
```

```python
res = []

for row in merged['price_diff']:
    if row > 0:
        res.append(1)
    else:
        res.append(0)

merged['inc_dec'] = res
merged['inc_dec'].value_counts()
merged.head()
```

```python
merged.drop(['High', 'Low', 'Volume', 'Adj Close', 'price_diff'], axis=1, inplace=True)
merged.head(10)
```

```python
merged.to_csv('dataset_with_target.csv')
```
